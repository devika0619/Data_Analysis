# 3. Which country relasenmore shows in netfilx

```python
data['country'].unique()
```

```python
# to remove the neasted  country and null values
df=data.copy()
df['country']=df['country'].str.split(',').explode('country')
df['country']=df['country'].str.lstrip()
df.replace([''],np.nan,inplace=True)
df.dropna(subset=['country'],inplace=True)
df
```

```python
# unique country
df['country'].describe(include='object')
```

```python
df['country'].nunique()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x='country',data=df,order=df['country'].value_counts().index[:10])
plt.xticks(rotation=90)
plt.show()
```

```python
# BOTTOM count_values(10 countries)
plt.figure(figsize=(5,5))
sns.countplot(x='country',data=df,order=df['country'].value_counts().index[-10:])
plt.xticks(rotation=90)
plt.show()
```

# 6.Which director produced more in netflix

```python
dg=data.copy()
dg['director']=dg['director'].str.split(',').explode('director')
dg['director']=dg['director'].str.lstrip()
dg.replace([''],np.nan,inplace=True)
dg.dropna(subset=['director'],inplace=True)
dg
```

```python
# 7.

t=a.assign(director=a['director'].str.split(',')).explode('director')
t['director']=t['director'].str.lstrip()
t.dropna(subset=['director'],inplace=True)
t
```

```python
t['director'].nunique()
```

```python
# Get the top 10 directors by count
top_10_directors = t['director'].value_counts().nlargest(10).index

# Filter the DataFrame to include only the rows with top 10 directors
filtered_data = t[t['director'].isin(top_10_directors)]

# Create the countplot
plt.figure(figsize=(10,5))
sns.countplot(x='director', data=filtered_data, order=top_10_directors)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Display the grid
plt.grid(True)

# Show the plot
plt.show()
```
